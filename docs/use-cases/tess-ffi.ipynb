```python
import numpy as np
import matplotlib.pyplot as plt
import lamatrix as la
import pickle
```

```python
res = pickle.load(open('data/tess_ffi.p', 'rb'))
```

```python
flux, flux_err, col, row, R, C, bmag = res['flux'], res['flux_err'], res['col'], res['row'], res['R'], res['C'], res['bmag']
```

```python
pix_mask = np.zeros(flux.shape, bool)
for r, c in zip(row, col):
    pix_mask |= np.hypot(R - r, C - c) < 5
```

```python
plt.pcolormesh(C, R, flux, vmin=0, vmax=1000)
plt.scatter(col, row, s=1, c='r')
plt.gca().set_aspect('equal')
```
